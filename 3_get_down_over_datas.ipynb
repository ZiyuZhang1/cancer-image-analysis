```python
path = '/A_LUAD/project/final_project/data/raw_data'
imgdir = os.listdir(path)
new_root = '/A_LUAD/project/final_project/data/nest_data'
for index, imgname in enumerate(imgdir):
  casename = imgname.split('_')[0]
  casepath = os.path.join(new_root, casename)
  oldpath = os.path.join(path, imgname)
  if not os.path.exists(casepath):
   os.makedirs(casepath)
  newpath = os.path.join(casepath, imgname)
  shutil.move(oldpath, newpath)
  # print(newpath,oldpath)
  # break
  
```

# get all cases need to be processed

```python
data = pd.read_csv('/A_LUAD/project/final_project/data/case_info.csv')
over_50_caselist = data['case_submitter_id'][data['image_num'] > 50].values
under_16_caselist = data['case_submitter_id'][data['image_num'] <= 16].values
```

# get downsample train,val,test data

## downsample train

```python
use_data = pd.read_csv('/A_LUAD/project/final_project/data/temp/two_train_images.csv')
```

```python
use_cases = set(use_data['case_submitter_id'].tolist())
```

```python
down_cases = use_cases.intersection(over_50_caselist)
```

```python
## undersample cases
all_path = '/A_LUAD/project/final_project/data/nest_data'
caselist = list(down_cases)
down_img_list = []
for data_case in caselist:
  casepath = os.path.join(all_path,data_case)
  imglist = os.listdir(casepath)
  selectimg = sample(imglist,50)
  for img in selectimg:
    imgname = img[:-4]
    down_img_list.append(imgname)
  imglist = []
```

```python
down_img_data = pd.DataFrame()
down_img_data['image_id'] = down_img_list
down_img_data = down_img_data.merge(use_data,how='inner',on='image_id')
```

```python
# no_change_list = list(use_cases-down_cases)
# def delete(use_data):
#    return row[0]+row[1]+row[2]

# df['new_col'] = df.apply(add, axis=1)
remove_list = []
idx = -1
for data_case in use_data.case_submitter_id:
  idx += 1
  if data_case in down_cases:
    remove_list.append(idx)
use_data.drop(remove_list, inplace=True)
```

```python
use_data = pd.concat([use_data,down_img_data])
use_data.to_csv('/A_LUAD/project/final_project/data/temp/down_train_images.csv',index=False)
```

# oversample images

```python
augcase = under_16_caselist
```

```python
## 先把所有需要扩增的文件夹单独取出等待处理
newroot = '/A_LUAD/project/final_project/data/temp/temp_aug_folder'
oldroot = '/A_LUAD/project/final_project/data/nest_data'
for case in augcase:
  oldcasedir = os.path.join(oldroot,case)
  newcasedir = os.path.join(newroot,case)
  os.makedirs(newcasedir)
  for img in os.listdir(oldcasedir):
    oldimgpath = os.path.join(oldcasedir,img)
    newimgpath = os.path.join(newcasedir,img)
    shutil.copy(oldimgpath,newimgpath)
```

```python
# 扩增取出的文件夹
root = '/A_LUAD/project/final_project/data/temp/temp_aug_folder'
for case in augcase:
  casepath = os.path.join(root,case)
  samplenum=len(os.listdir(casepath)*3)
  # print(samplenum,case)
  p = Augmentor.Pipeline(casepath)
  # p.rotate90(probability=0.5)
  p.rotate270(probability=0.5)
  p.flip_left_right(probability=0.8)
  p.flip_top_bottom(probability=0.3)
  p.sample(samplenum)
```

```python
## 释放子文件夹复制的图片
root = '/A_LUAD/project/final_project/data/temp/temp_aug_folder'
for folder in os.listdir(root):
  folderpath = os.path.join(root,folder)
  subfolderpath = os.path.join(folderpath,'output')
  for img in os.listdir(subfolderpath):
    oldimgpath = os.path.join(subfolderpath,img)
    newimgpath = os.path.join(folderpath,img)
    shutil.move(oldimgpath,newimgpath)
  os.removedirs(subfolderpath)
```

```python
## 把不需要改变的case folder加到目前只有aug的文件夹里
allcaselist = data['case_submitter_id'].tolist()
nochangelist = set(allcaselist)-set(augcase)
nochangelist = list(nochangelist)
newroot = '/A_LUAD/project/final_project/data/temp/temp_aug_folder'
oldroot = '/A_LUAD/project/final_project/data/nest_data'
for caseid in nochangelist:
  oldcasepath = os.path.join(oldroot,caseid)
  newcasepath = os.path.join(newroot,caseid)
  # os.mkdir(newcasepath)
  for img in os.listdir(oldcasepath):
    oldimgpath = os.path.join(oldcasepath,img)
    newimgpath = os.path.join(newcasepath,img)
    shutil.copy(oldimgpath,newimgpath)
    # print(oldimgpath,newimgpath)
```
