```python
import glob
import numpy as np
import pickle

import tensorflow as tf
from tensorflow import keras
import pandas as pd
import os
from tensorflow.keras.callbacks import TensorBoard
import time
import pickle
import matplotlib.pyplot as plt
```

# get train and val data

```python
import glob
group1 = glob.glob('/A_LUAD/project/final_project/data/temp/two_train/group1/*.png')
group2 = glob.glob('/A_LUAD/project/final_project/data/temp/two_train/group2/*.png')

data = []
labels = []

for i in group1:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (300,300))
    image=np.array(image)
    data.append(image)
    labels.append(0)
for i in group2:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (300,300))
    image=np.array(image)
    data.append(image)
    labels.append(1)

X_train = np.array(data)
y_train = np.array(labels)
```

```python
import glob
group1 = glob.glob('/A_LUAD/project/final_project/data/temp/two_val/group1/*.png')
group2 = glob.glob('/A_LUAD/project/final_project/data/temp/two_val/group2/*.png')

data = []
labels = []

for i in group1:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (300,300))
    image=np.array(image)
    data.append(image)
    labels.append(0)
for i in group2:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (300,300))
    image=np.array(image)
    data.append(image)
    labels.append(1)

X_val = np.array(data)
y_val = np.array(labels)
```

# pickle

```python
import pickle
drive_path = '/A_LUAD/project/final_project/data/temp'
# Dump the data into a pickle file
with open(drive_path + '/X_train.pickle', 'wb') as f:
    pickle.dump(X_train, f)
with open(drive_path + '/y_train.pickle', 'wb') as f:
    pickle.dump(y_train, f)
with open(drive_path + '/X_val.pickle', 'wb') as f:
    pickle.dump(X_val, f)
with open(drive_path + '/y_val.pickle', 'wb') as f:
    pickle.dump(y_val, f)
```
